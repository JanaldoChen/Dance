```python
import os
import torch
import trimesh
from PIL import Image
import cv2
```

```python
data_root = 'data/people_snapshot_public'
people_ID_list = os.listdir(data_root)
print(len(people_ID_list), people_ID_list)
```

```python
def video_to_images(vid_path, imgs_path):
    cap = cv2.VideoCapture(vid_path)
    i = 0

    if not os.path.exists(imgs_path):
        os.makedirs(imgs_path)
    else:
        for fn in os.listdir(imgs_path):
            os.remove(os.path.join(imgs_path, fn))
        
    while(cap.isOpened()):
        flag, frame = cap.read()
        if flag == False:
            break
        img_fn =  '{:0>5d}.png'.format(i)
        cv2.imwrite(os.path.join(imgs_path, img_fn), frame)
        i += 1
    print('Total frames: ', i)
```

```python
for people_ID in people_ID_list:
    video_to_images(vid_path=os.path.join(data_root, people_ID, people_ID+'.mp4'), imgs_path=os.path.join(data_root, people_ID, 'images'))
```

```python
data_root = 'data/iPER'
video_ID_list = os.listdir(os.path.join(data_root, 'iPER_256_video_release'))
print(len(video_ID_list), video_ID_list)
```

```python
for video_ID in video_ID_list:
    img_paths = video_ID[:-4].split('_')
    imgs_path=os.path.join(data_root, 'images', img_paths[0], img_paths[1], img_paths[2])
    print(imgs_path)
    video_to_images(vid_path=os.path.join(data_root, 'iPER_256_video_release', video_ID), imgs_path=imgs_path)
```
